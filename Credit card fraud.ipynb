##### What is a PCA?

Principal component analysis (PCA) is a linear dimensionality reduction technique.

The data is linearly transformed onto a new coordinate system such that the directions (principal components) capturing the largest variation in the data can be easily identified. 

The principal components of a collection of points in a real coordinate space are a sequence of p unit vectors, where the i-th vector is the direction of a line that best fits the data while being orthogonal to the first i âˆ’ 1 vectors. Here, a best-fitting line is defined as one that minimizes the average squared perpendicular distance from the points to the line. These directions (i.e., principal components) constitute an orthonormal basis in which different individual dimensions of the data are linearly uncorrelated. Many studies use the first two principal components in order to plot the data in two dimensions and to visually identify clusters of closely related data points. 

##### PCA in this dataframe

<span style="color:MediumVioletRed"> Data Preprocessing: </span> The original dataset likely contained a variety of features related to credit card transactions, such as transaction amount, time of transaction, merchant category, etc.

<span style="color:MediumVioletRed"> Anonymization: </span> To protect the privacy of individuals and businesses, any personally identifiable information or sensitive details are removed or anonymized. This is crucial for compliance with data protection regulations like GDPR.

<span style="color:MediumVioletRed">Dimensionality Reduction with PCA: </span> After anonymization, PCA (Principal Component Analysis) is applied to the dataset. PCA is a technique used for dimensionality reduction, which transforms the original features into a new set of orthogonal variables called principal components.

<span style="color:MediumVioletRed"> Principal Components (V1-V28): </span> The resulting principal components (V1-V28) are linear combinations of the original features. Each principal component captures different patterns and relationships within the data while maximizing the explained variance. The components are ordered by the amount of variance they explain, with V1 explaining the most variance and subsequent components explaining progressively less.

<span style="color:MediumVioletRed"> Selection of Principal Components: </span> Typically, a subset of the principal components that captures most of the variance in the data is selected for further analysis. These selected components (V1-V28) serve as the input features for credit card fraud detection algorithms.

```python
print("Histograms")
num_columns = len(df.columns)
num_rows = (num_columns + 3) // 4 

fig, axes = plt.subplots(num_rows, 4, figsize=(16, 4*num_rows))

if not isinstance(axes, np.ndarray):
    axes = np.array(axes)

for i, column in enumerate(df.columns):
    ax = axes[i // 4, i % 4]  
    ax.hist(df[column], bins=50, color='lightgreen', alpha=0.7)
    ax.set_title(f'Histogram of {column}')
    ax.set_xlabel('Values')
    ax.set_ylabel('Frequency')
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
print("Box plots")
fig, axes = plt.subplots(num_rows, 4, figsize=(16, 4*num_rows))
if not isinstance(axes, np.ndarray):
    axes = np.array(axes)

for i, column in enumerate(df.columns):
    ax = axes[i // 4, i % 4]  
    sns.boxplot(x=column, data=df, color='skyblue', ax=ax)
    ax.set_title(f'Box Plot of {column}')
    ax.set_xlabel(column)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(num_rows, 4, figsize=(16, 4*num_rows))
if not isinstance(axes, np.ndarray):
    axes = np.array(axes)

for i, column in enumerate(df.columns):
    ax = axes[i // 4, i % 4]  
    sns.kdeplot(df[column], shade=True, color='teal', ax=ax)
    ax.set_title(f'KDE Plot of {column}')
    ax.set_xlabel(column)
    ax.set_ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(num_rows, 4, figsize=(16, 4*num_rows))
if not isinstance(axes, np.ndarray):
    axes = np.array(axes)

for i, column in enumerate(df.columns):
    ax = axes[i // 4, i % 4] 
    sns.violinplot(x=column, data=df, color='coral', ax=ax)
    ax.set_title(f'Violin Plot of {column}')
    ax.set_xlabel(column)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(num_rows, 4, figsize=(16, 4*num_rows))
if not isinstance(axes, np.ndarray):
    axes = np.array(axes)

for i, column in enumerate(df.columns):
    ax = axes[i // 4, i % 4]  # Get the current axis
    sns.ecdfplot(data=df, x=column, ax=ax)
    ax.set_title(f'CDF Plot of {column}')
    ax.set_xlabel(column)
    ax.set_ylabel('Cumulative Probability')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(num_rows, 4, figsize=(16, 4*num_rows))
if not isinstance(axes, np.ndarray):
    axes = np.array(axes)

for i, column in enumerate(df.columns):
    ax = axes[i // 4, i % 4]  # Get the current axis
    stats.probplot(df[column], dist="norm", plot=ax)
    ax.set_title(f'QQ Plot of {column}')
    ax.set_xlabel('Theoretical Quantiles')
    ax.set_ylabel('Ordered Values')

plt.tight_layout()
plt.show()
```
